```python
import numpy as np
import tensorflow as tf
import gym
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def stats(all_rewards):
    print("Mean of all rewards: ", np.mean(all_rewards))
    all_rewards_100 = []
    for i in range(100,num_episodes + 1):
        all_rewards_100.append(np.mean(all_rewards[i-100:i]))
    print("Max 100 rewards average: ", np.max(all_rewards_100))
    me = np.argmax(all_rewards_100)
    print("From episode: %d to episode %d" % (me, me + 99))
    plt.plot(range(len(all_rewards_100)), all_rewards_100)
```

# Q-Learning with lookup table

```python
env = gym.make('FrozenLake-v0')
num_episodes = 5000

learning_rate = 0.05

Q = np.zeros((16,4))
epsilon = 0.001
all_rewards = []
for episode in range(num_episodes):
    s = env.reset()
    done = False
    while not done:
        if np.random.random_sample() < epsilon:
            a = env.action_space.sample()
        else:
            a = np.argmax(Q[s])
        s_prime, reward, done, _ = env.step(a)
    
        if done and reward == 0.0:
            reward = -1.0

        Q[s][a] += learning_rate * (reward + np.max(Q[s_prime]) - Q[s][a])
        s = s_prime
    
    all_rewards.append(reward == 1)
    
env.close()
stats(all_rewards)
```

# Q-Learning with function approximation

```python
env = gym.make('FrozenLake-v0')
num_episodes = 8001
nnn = 500

learning_rate = 0.1

tf.reset_default_graph()
X = tf.placeholder(dtype=tf.float32, shape=(None, 4 * 4))
y = tf.placeholder(dtype=tf.float32, shape=(None, 4))
W = tf.Variable(initial_value=tf.random_uniform((16, 4),0,0.01))
Q = tf.matmul(X, W)
mse = tf.reduce_sum(tf.square(y - Q))
train_op = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(mse)

sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
id16 = np.identity(16)
all_rewards = []
for episode in range(num_episodes):
    s = env.reset()
    done = False
    i=0
    epsilon = 1./((episode/50) + 10)
    while not done:
        Q_s_A = Q.eval(feed_dict={X: [id16[s]]})
        if np.random.random_sample() < epsilon:
            a = env.action_space.sample()
        else:
            a = np.argmax(Q_s_A[0])
        s_prime, reward, done, _ = env.step(a)
        i+=reward
        if done:
            all_rewards.append(reward)
        if done and reward < 0.5:
            reward = -1.0
        Q_s_prime_a_prime = Q.eval(feed_dict={X: [id16[s_prime]]})
        if np.random.random_sample() < epsilon:
            Q_s_prime_a_prime = Q_s_prime_a_prime[0][env.action_space.sample()]
        else:
            Q_s_prime_a_prime = np.max(Q_s_prime_a_prime)
        
        Q_s_A[0][a] = reward + 0.99*Q_s_prime_a_prime
        
        train_op.run(feed_dict={X: [id16[s]], y: Q_s_A})
        s = s_prime
    if episode % nnn == 0:# or True:# or True:
        print(episode, reward)
env.close()
stats(all_rewards)
```
